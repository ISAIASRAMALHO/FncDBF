```python
from dbfread import DBF
import sys
import glob
from pandas import DataFrame
```

```python
tabela = DBF('A_CLIENT.DBF', encoding='latin1')
frame = DataFrame( tabela )
frame
```

```python
cmp_lista = frame['C_CLIENTE']
```

```python
def descript( v ):
    formula = 0
    r = ''
    caractere = ''
    for i in range(0, len( v ) ):
        formula = ord( int( v[i] ) - ( 22 ) * (i + 1) )
        caractere = chr( formula )
        r += caractere
    return r    


for c in cmp_lista:
    print( descript( c ) )
```
